```
import re
import regex
import unittest 

import pandas
```

```
def find_pattern(pattern, string):
    pattern = re.compile(pattern)
    if pattern.search(string) != None:
        return True
    else:
        return False
```

## 1. Underscore

Implemente uma função usando expressões regulares que teste se uma string contém uma sequência de letras minúsculas seguidas por um `underscore`. Mostre exemplos.



```
lowercase_and_underscore = '[a-z]+_.*'
```

#### Casos de teste

```
class TestUnderscore(unittest.TestCase):  
    def test_correct(self):
        self.assertTrue(find_pattern(lowercase_and_underscore, 'asdafaf_asd'))

    def test_lowercase_uppercase_lowercase_underscore(self):
        self.assertTrue(find_pattern(lowercase_and_underscore, 'asdAdasda_'))

    def test_lowercase_digit_lowercase_underscore(self):
        self.assertTrue(find_pattern(lowercase_and_underscore, 'sadgfw9fdad_'))
    
    def test_underscore(self):
        self.assertFalse(find_pattern(lowercase_and_underscore, '_'))
    
    def test_lowercase(self):
        self.assertFalse(find_pattern(lowercase_and_underscore, 'asdadasd'))
    
    def test_underscore_uppercase_underscore(self):
        self.assertFalse(find_pattern(lowercase_and_underscore, '___A___'))
    
    def test_underscore_lowercase(self):
        self.assertFalse(find_pattern(lowercase_and_underscore, '___asdasdda'))
    
    def test_digit(self):
        self.assertFalse(find_pattern(lowercase_and_underscore, '1231'))
    
    def test_punctuation_underscore(self):
        self.assertFalse(find_pattern(lowercase_and_underscore, '.;,._'))
```

```
class TestCpf(unittest.TestCase):
    def test_correct_format(self):
        self.assertTrue(find_pattern(cpf_pattern, '123.456.789-09'))

    def test_incorrect_format(self):
        self.assertFalse(find_pattern(cpf_pattern, '12345678909'))

    def test_empty_string(self):
        self.assertFalse(find_pattern(cpf_pattern, ''))

    def test_cpf_with_letters(self):
        self.assertFalse(find_pattern(cpf_pattern, 'abc.def.ghi-jk'))
    
    def test_comma_instead_of_period(self):
        self.assertFalse(find_pattern(cpf_pattern, '123,465,789-09'))

    def test_underscore_instead_of_hyphen(self):
        self.assertFalse(find_pattern(cpf_pattern, '123.465.789_09'))
```

## 3. Endereços

Implemente uma função usando expressões regulares que seja capaz, com uma única expressão regular, de extrair rua, número, apartamento (opcionalmente), bairro (opcionalmente), CEP, cidade e estado. Esta função deve ser capaz de funcionar em todos os exemplos abaixo:


*   Rua José da Silva, 346, Farol, CEP 57002-220. Maceió, AL.
*   Rua da Consolação, 9999, apt 302, CEP 11022-202. São Paulo, SP.
*   Avenida Atlântica, 420, ap 1001, Copacabana, cep 22011-010. Rio de Janeiro, RJ.



```
address_pattern = '(?P<Rua>[A-Za-záàãâéêíóõúç\s]+), ' + \
                  '?(?P<Número>\d*), ' + \
                  '(apt? (?P<Apartamento>\d+), )?' + \
                  '((?P<Bairro>[A-Z][a-z]+), )?' + \
                  '((cep|CEP) (?P<CEP>\d{5}\-\d{3})). ' + \
                  '(?P<Cidade>[A-Za-záàãâéêíóõúç\s]+), ' + \
                  '(?P<Estado>[A-Z]{2})'


address_list = ['Rua José da Silva, 346, Farol, CEP 57002-220. Maceió, AL.',
                'Rua da Consolação, 9999, apt 302, CEP 11022-202. São Paulo, SP.',
                'Avenida Atlântica, 420, ap 1001, Copacabana, cep 22011-010. Rio de Janeiro, RJ.']
rows = []

for address in address_list:
    rows.append(re.match(address_pattern, address).groupdict())

df = pandas.DataFrame(data=rows).fillna('-')
display(df)
```

#### Casos de teste

```
class TestAddress(unittest.TestCase):
    def test_first_address(self):
        self.assertEqual(dict(re.match(address_pattern, address_list[0]).groupdict()), {
            'Rua': 'Rua José da Silva',
            'Número': '346',
            'Apartamento': None,
            'Bairro': 'Farol',
            'CEP': '57002-220',
            'Cidade': 'Maceió',
            'Estado': 'AL',
        })

    def test_second_address(self):
        self.assertEqual(dict(re.match(address_pattern, address_list[1]).groupdict()), {
            'Rua': 'Rua da Consolação',
            'Número': '9999',
            'Apartamento': '302',
            'Bairro': None,
            'CEP': '11022-202',
            'Cidade': 'São Paulo',
            'Estado': 'SP',
        })

    def test_third_address(self):
        self.assertEqual(dict(re.match(address_pattern, address_list[2]).groupdict()), {
            'Rua': 'Avenida Atlântica',
            'Número': '420',
            'Apartamento': '1001',
            'Bairro': 'Copacabana',
            'CEP': '22011-010',
            'Cidade': 'Rio de Janeiro',
            'Estado': 'RJ',
        })
```

## 4. URLs

Implemente uma função usando expressões regulares que encontre todas as URLs em uma string. Mostre exemplos

```
scheme = '([A-Za-z]+://)?'
host = '([\w][\w\-\.]+)'
port = '(:\d{1,5})?'
domain = '(\.[A-Za-z]+)*'
# Lista de caracteres permitidos no path da URL de acordo com o RFC 3986
# https://stackoverflow.com/a/4669750
path = '(/[\w\.\-_~!$&\'()\*\+,;=:@%])*' 
query = '(\?|\?([\w\.\-_~!$&\'()\*\+,;=:@%]+=?[\w\.\-_~!$&\'()\*\+,;=:@%]*)+)?'
fragment = '(#|#[\w\.\-_~!$&\'()\*\+,;=:@%])?'

url_pattern = f'{scheme}(www\.)?{host}{port}{domain}{path}{query}{fragment}'
```

```
def find_urls(string):
    pattern = re.compile(url_pattern)
    return pattern.findall(string)
```

#### Casos de teste

```
class TestUrl(unittest.TestCase):
    def test_valid_url(self):
        self.assertTrue(bool(find_urls('www.google.com.br')))
    
    def test_invalid_url(self):
        self.assertFalse(bool(find_urls('')))

    def test_starts_with_url(self):
        self.assertTrue(bool(find_urls('www.google.com.br é o site mais visitado do mundo')))

    def test_ends_with_url(self):
        self.assertTrue(bool(find_urls('Para saber mais, visite www.google.com.br')))
    
    def test_multiple_url(self):
        self.assertTrue(bool(find_urls('Para saber mais, visite www.google.com.br ou valer.io')))

    def test_classroom_url(self):
        self.assertTrue(bool(find_urls('https://classroom.google.com/c/NjI1Mjc5NTg1MzBa/a/MTM0OTYwMzA3NTg3/details')))

    def test_url_with_port(self):
        self.assertTrue(bool(find_urls('localhost:8000')))

    def test_mathias_bynens(self):
        urls = '''http://foo.com/blah_blah
http://foo.com/blah_blah/
http://foo.com/blah_blah_(wikipedia)
http://foo.com/blah_blah_(wikipedia)_(again)
http://www.example.com/wpstyle/?p=364
https://www.example.com/foo/?bar=baz&inga=42&quux
http://userid:password@example.com:8080
http://userid:password@example.com:8080/
http://userid@example.com
http://userid@example.com/
http://userid@example.com:8080
http://userid@example.com:8080/
http://userid:password@example.com
http://userid:password@example.com/
http://142.42.1.1/
http://142.42.1.1:8080/
http://foo.com/blah_(wikipedia)#cite-1
http://foo.com/blah_(wikipedia)_blah#cite-1
http://foo.com/(something)?after=parens
http://code.google.com/events/#&product=browser
http://j.mp
ftp://foo.bar/baz
http://foo.bar/?q=Test%20URL-encoded%20stuff
http://-.~_!$&'()*+,;=:%40:80%2f::::::@example.com
http://1337.net
http://a.b-c.de
http://223.255.255.254'''.split('\n')

        for url in urls:
            self.assertTrue(bool(find_urls(url)))
    

  
```

## 5. Citações

Considere o seguinte exemplo de referência de livro em formato de citação APA:

> _Manning, C. D., & Schutze, H. (1999). Foundations of statistical natural language processing. MIT press._

Implemente uma função usando expressões regulares que extraia cada autor, ano de publicação, título e editora do livro, para qualquer referência de livro neste formato.

```
def segmentate_citation(citation):
    '''
    Extrai informação (lista de autores, ano de publicação, título e editora) de
    citações no formato APA. Se a citação for válida então retorna um dicionário
    contendo as informações, caso contrário retorna um dicionário
    '''
    
    initial = '[A-Z]\.'
    surname = '[A-Z][a-z]+'
    author = f'(?P<Autor>{surname},?(?:\s{initial})+)'
    author_list = f'({author})(((?:, ({author}))*(?:, & ({author})))|())'
    year = '(?P<Ano>\d+)'
    title = '(?P<Título>[\w,\-\s:,()]+)'
    pub = '(?P<Editora>[\w\s:,]+)'

    citation_pattern = regex.compile(f'{author_list} \({year}\)\. {title}\. {pub}\.')
    match = citation_pattern.match(citation)
    return match.capturesdict() if match else {}
```

#### Casos de teste



```
class TestCitations(unittest.TestCase):
    def test_empty_string(self):
        self.assertEqual(segmentate_citation(''), {})

    def test_example(self):
        citation = 'Manning, C. D., & Schutze, H. (1999). Foundations of statistical natural language processing. MIT press.'
        expected = {
            'Autor': ['Manning, C. D.', 'Schutze, H.'],
            'Ano': ['1999'],
            'Título': ['Foundations of statistical natural language processing'],
            'Editora': ['MIT press']
        }

        self.assertEqual(segmentate_citation(citation), expected)

    def test_author_without_initials(self):
        citation = 'Manning, & Schutze, H. (1999). Foundations of statistical natural language processing. MIT press.'
        expected = {}

        self.assertEqual(segmentate_citation(citation), expected)

    def test_last_author_without_and(self):
        citation = 'Manning, C. D., Schutze, H. (1999). Foundations of statistical natural language processing. MIT press.'
        expected = {}

        self.assertEqual(segmentate_citation(citation), expected)

    def test_year_without_parenthesis(self):
        citation = 'Manning, C. D., & Schutze, H. 1999. Foundations of statistical natural language processing. MIT press.'
        expected = {}
        
        self.assertEqual(segmentate_citation(citation), expected)

    def test_only_one_author(self):
        citation = 'Manning, C. D. (1999). Foundations of statistical natural language processing. MIT press.'
        expected = {
            'Autor': ['Manning, C. D.'],
            'Ano': ['1999'],
            'Título': ['Foundations of statistical natural language processing'],
            'Editora': ['MIT press']
        }

        self.assertEqual(segmentate_citation(citation), expected)

    def test_three_authors(self):
        citation = 'Silva, A., Silva, B., Silva, C., & Silva, D. (1999). Livro fictício. Editora mais fictícia ainda.'
        expected = {
            'Autor': ['Silva, A.', 'Silva, B.', 'Silva, C.', 'Silva, D.'],
            'Ano': ['1999'],
            'Título': ['Livro fictício'],
            'Editora': ['Editora mais fictícia ainda']
        }

        self.assertEqual(segmentate_citation(citation), expected)

    def test_valid_citations(self):
        '''
        Retirados em : https://apastyle.apa.org/style-grammar-guidelines/references/examples/book-references
        '''
        citations = [
             'Jackson, L. M. (2019). The psychology of prejudice: From attitudes to social action (2nd ed). American Psychological Association.',
             'Sapolsky, R. M. (2017). Behave: The biology of humans at our best and worst. Penguin Books.',
             'Watson, J. B., & Rayner, R. (2013). Conditioned emotional reactions: The case of Little Albert (D Webb, Ed). CreateSpace Independent Publishing Platform.'
        ]
        expected = [
            {
                'Autor': ['Jackson, L. M.'],
                'Ano': ['2019'],
                'Título': ['The psychology of prejudice: From attitudes to social action (2nd ed)'],
                'Editora': ['American Psychological Association'],
            },
            {
                'Autor': ['Sapolsky, R. M.'],
                'Ano': ['2017'],
                'Título': ['Behave: The biology of humans at our best and worst'],
                'Editora': ['Penguin Books'],
            },
            {
                'Autor': ['Watson, J. B.', 'Rayner, R.'],
                'Ano': ['2013'],
                'Título': ['Conditioned emotional reactions: The case of Little Albert (D Webb, Ed)'],
                'Editora': ['CreateSpace Independent Publishing Platform'],
            }
        ]
        
        self.assertEqual(list(map(segmentate_citation, citations)), expected)


```
